# Surprise Housing Dataset

A US-based housing company named Surprise Housing has decided to enter the Australian market. The company uses data analytics to purchase houses at a price below their actual values and flip them on at a higher price. For the same purpose, the company has collected a data set from the sale of houses in Australia .The company is looking at prospective properties to buy to enter the market.

```python
# checking all the missing values
miss_val = (df.isnull().sum() / df.shape[0]) * 100
miss_val
```

```python
# dropping all the columns having missing value % > 17
rem_cols = ['Id','LotFrontage','LotFrontage','Fence','PoolQC','FireplaceQu','LotFrontage','Alley','MiscFeature','GarageYrBlt','MasVnrArea']

df.drop(rem_cols,axis = 1,inplace= True)
```

```python
# Checking all the missing values
miss_val = (df.isnull().sum() / df.shape[0]) * 100
miss_val
```

```python
# Selecting all the categorical features 
cat_list = df.select_dtypes(include = 'object')
cat_list.head()
```

```python
# Checking the missing values
cat_miss = cat_list.isnull().sum()
cat_miss = cat_miss.sort_values(ascending = False)
cat_miss
```

```python
df[cat_miss.index].head()
```

```python
# Using SimpleImputer to replace all the nan values with the most frequent occurance 

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = 'most_frequent')

df[cat_miss.index] = imputer.fit_transform(df[cat_miss.index])
df[cat_miss.index].isnull().sum()
```

```python
# using label encoder to encode all the categories into respected numerical form

from sklearn.preprocessing import LabelEncoder

cat = cat_miss.index

le = LabelEncoder()

df[cat] = df[cat].astype('category')

df[cat] = le.fit_transform(cat)

df[cat].head()
```

```python
# Selecting the integer datatype

int_list = df.select_dtypes(include=['float64','int64'])
int_list.head()
```

```python
# Checking the missing values for the numerical data

int_list.isnull().sum()
```

```python
# scaling the integer variables

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[int_list.columns] = scaler.fit_transform(df[int_list.columns])
df[int_list.columns].head()
```

```python
# Checking if the data is properly scaled or not
df[int_list.columns].describe()
```

```python
# now looking at the final dataset

df.head()
```

```python
# Creating X & y

X = df.drop('SalePrice',1)
y = df['SalePrice']

X.shape,y.shape
```

```python
# using ExtraTreesRegressor to retrieve best features from the dataset

from sklearn.ensemble import ExtraTreesRegressor

etr = ExtraTreesRegressor()
etr.fit(X,y)
```

```python
# Retriving top n features from the df
top_features = pd.Series(etr.feature_importances_,index=X.columns)
top_features = top_features.nlargest(15)
top_features
```

```python
# Performing train_test_split
from sklearn.model_selection import train_test_split

X_train, X_test ,y_train ,y_test = train_test_split(X,y , train_size = 0.7 ,random_state = 102)
X_train.shape, X_test.shape ,y_train.shape ,y_test.shape
```

Before starting to build a model for the dataset using LazyPredict library to get the best machine learning model

## RandomForestRegressor 

```python
# RandomForestRegressor

from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth = 5 , n_estimators = 100 ,random_state=100)

rf.fit(X_train,y_train)

rf.feature_importances_
```

```python
#checking the top 15 features according to randomforest
top_15 = pd.Series(rf.feature_importances_,index =X_train.columns).nlargest(15)
top_15
```

```python
print("Training accuracy = {}".format(rf.score(X_train,y_train)))
print("Test accuracy = {}".format(rf.score(X_test,y_test)))
```

```python
#Tuning the hyperparameters 
from sklearn.model_selection import RandomizedSearchCV

#random forest model to be hypertuned
rf1 = RandomForestRegressor()

#parameter grid 
params = {
    'criterion' : ['mse','mae'],
    'bootstrap' : 'True',
    'max_features': ['auto', 'sqrt'],
    'n_estimators' : [10, 20 ,30 ,50 ,100, 200 ,400 , 600 ],
    'max_depth' : [3,4,5,6,7,8,9,10,12,15],
    'min_samples_leaf' : [2,3,4,5,6,7,8 ],
    'min_samples_split' :[2, 5 , 10 ,20 ]
}

rcv = RandomizedSearchCV(cv = 10 ,n_jobs = -1 , estimator = rf1 ,n_iter= 10, random_state = 102 ,param_distributions =  params, return_train_score = True)

rcv.fit(X_train ,y_train)
```

```python
#best parameter from randomsearchcv
rcv.best_params_
```

```python
rcv.best_estimator_
```

```python
rcv.best_score_
```

```python
# Building a RandomForestRegressor with hypertuned parameters

RandomForest = RandomForestRegressor(bootstrap=True, criterion = 'mse',max_features= 'auto' , max_depth=15, min_samples_leaf=3,min_samples_split=5, n_estimators=50)

RandomForest.fit(X_train,y_train)
```

```python
print("Training accuracy = {}".format(RandomForest.score(X_train,y_train)))
print("Test accuracy = {}".format(RandomForest.score(X_test,y_test)))
```

```python
#checking the top 15 features according to randomforest
top_15_rf = pd.Series(RandomForest.feature_importances_,index =X_train.columns).nlargest(15)
top_15_rf.plot(kind = 'barh')
```
